```python
import time
from dask.distributed import Client, wait

def process_num(x):
    """
    A simple function that processes a number.
    Instead of printing, it returns the result.
    """
    # You could do complex work here, e.g., return x * 2
    return x

# THIS IS THE MOST IMPORTANT PART!
# Always protect your Dask Client and task submission
# inside this 'if' block.
if __name__ == '__main__':
    
    # 1. Start a local Dask cluster with 20 workers
    client = Client("tls://192.168.115.147:8786")
    
    print(f"Dask Dashboard link: {client.dashboard_link}")
    print("--- Submitting 100 tasks to 20 workers ---")

    futures = []
    # 2. Create tasks for numbers 1 to 100
    for i in range(1, 101):
        future = client.submit(process_num, i)
        futures.append(future)

    # 3. Wait for all tasks to finish
    wait(futures)
    print("--- All tasks complete. Gathering results... ---")

    # 4. Gather the results from all workers back to the main script
    results = client.gather(futures)
    
    # 5. Now, print the results from your main script
    print("--- Results ---")
    print(results)

    # 6. Shut down the cluster
    client.close()
```
